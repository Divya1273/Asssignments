# Import Libraries

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules,apriori
import warnings
warnings.filterwarnings('ignore')
```

# Load the data

```python
df = pd.read_csv('Online retail.csv', header=None)
df
```

```python
df.dtypes
```

```python
df.columns
```

```python
df.iloc[[0]]
```

```python
df.isnull().sum()
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
df.head()
```

```python
data = df[0].apply(txt_split)
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
data
```

```python
data.replace([True,False],[1,0], inplace=True)
```

```python
data
```

```python
data.shape
```

```python
scores = apriori(data, min_support=0.01, use_colnames=True)
scores
```

```python
rules = association_rules(scores, metric="confidence", min_threshold=0.2)
rules
```

```python
rules.sort_values(by='lift', ascending=False, inplace=True)
rules
```

```python
'''
Loaded transaction data, checked for null values, and encoded items into binary format using TransactionEncoder.
Applied the Apriori algorithm with a minimum support threshold to find commonly bought item combinations.
Used the association_rules() function to extract rules with metrics like support, confidence, and lift.
Items like herb & pepper and ground beef are commonly bought together, 
followed by combinations like mineral water, spaghetti, and ground beef.
'''
```

## Interview Questions

#### 1.	What is lift and why is it important in Association rules?

```python
'''
Lift in association rules measures how much more likely two items are to appear together compared to if they were independent.
A lift value greater than 1 means the items are positively related, while less than 1 means they occur less together than expected.
It's important because it helps identify strong relationships between items in data, which can better for decision-making.
'''
```

#### 2.	What is support and Confidence. How do you calculate them?

```python
'''
Support measures how frequently an item or itemset appears in the dataset.
Confidence tells you how likely item B is purchased when item A is purchased.
These metrics help in finding strong and relevant associations between items.
'''
```

#### 3.	What are some limitations or challenges of Association rules mining?

```python
'''
Some challenges of association rule mining are:
Scalability: It can be slow and resource-intensive with large datasets.
Too many rules: It can generate a lot of rules, many of which might not be useful.
Overfitting: Rules may fit the data too closely and not apply to other datasets.
'''
```
